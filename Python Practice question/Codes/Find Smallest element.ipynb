**Question-Find The Smallest Element:**

You have to find and print the smallest element of the list given as input. 

----------------------------------------------------------------------
Input:
A non-empty list of integers.

Output:
The smallest integer of the input list.

----------------------------------------------------------------------
Sample input:
[2, -3, 0, 7, 21]

Sample output:
-3

```python
import time

start=time.time()

input=[2, -3, 0, 7, 21]

b=input
c=b[0]
for i in range(len(input)):
  for j in b:
    if j<c:
      c=j

end=time.time()
print('smallest no. = ', c)
print('Time taken by code :' , end-start)
```
