**Fibonacci series**

Compute and display Fibonacci series upto n terms where n is a positive integer entered by the user.

Sample Input: 5

Sample Output:

0

1

1

2

3

```python
a=0
b=1

number=int(input('Enter how long Fibonacci series is needed = '))

print('s.no  -  Fibonacci series')
print('1    - ',a)
print('2    - ',b)
for i in range(2,number):
  c=a+b
  print(i+1,'   - ',c)
  a=b
  b=c
```
