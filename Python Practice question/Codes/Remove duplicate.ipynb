**Question-Remove duplicate:**

Write a code to remove duplicate elements from a list


----------------------------------------------------------------------

Input:

[12, 36, 56, 36, 36, 50, 56, 12] 


Output:

[12, 36, 56, 50] 


```python
# First Method - Through Loop

a= [12, 36, 56, 36, 36, 50, 56, 12]

b=[]

for i in range(len(a)):
  if a[i] not in b:
    b.append(a[i])
b
```

```python
# Second Method - Through converting it in set

a= [12, 36, 56, 36, 36, 50, 56, 12]

b=set(a)

print('convert to set- ',b)

b=list(b)

print('converted back to list - ' ,b)
```
