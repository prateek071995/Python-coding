**Pattern:**

Printing different patterns is a very good exercise to reinforce iteration through loops and strong logic building. Here you will be given a positive integer and you will generate pattern based on that integer.

----------------------------------------------------------------------
Input:
A positive integer n
1 <= n <=20

Output:
A pattern as described by the Sample input and outputs below.

----------------------------------------------------------------------
Sample input:
5

```python
#Pattern -1
a=int(input('enter no. of lines to print in pattern - '))

for i in range(1,a+1):
  print('*'*i)
```

```python
#Pattern -2
a=int(input('enter no. of lines to print in pattern - '))
c=a
for i in range(1,a+1):
  b=i-1
  print(' '*c,'*'*i,end='' )
  print('*'*b)
  c=c-1
```

```python
#Pattern -3
a=int(input('enter no. of lines to print in pattern - '))
b=0
for i in range(1,a+1):
  
  print(' '*a,'*_'*b,end='')
  print('*')
  b=b+1
  a=a-1
```

```python
# 2nd method
print('Pattern - 3')
a= int(5)

for i in range(0,a):
    print(' '*(a-i),end='')
    print('*_'*i,end = '')
    print('*')
```
